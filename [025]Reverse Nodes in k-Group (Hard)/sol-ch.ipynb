```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseKGroup(self, head, k):
        """
        :type head: ListNode
        :type k: int
        :rtype: ListNode
        """
        new_linked_list = ListNode()
        new_head = new_linked_list
        
        while check_k_node(head, k):
            k_node = get_k_node(head, k)
            new_head.next = reverse_linked_list(k_node)
            head = skip_k_nodes(head, k)
            new_head = skip_k_nodes(new_head, k)
            
        while head is not None:
            new_head.next = ListNode(head.val)
            new_head = new_head.next
            head = head.next
        
        print_linked_list(new_linked_list)
        new_linked_list = new_linked_list.next
        
        return new_linked_list
            
def check_k_node(head, k):
    while k > 0:
        if head is not None:
            head = head.next
            k -= 1
        else:
            return False
    return True

def get_k_node(head, k):
    new_linked_list = ListNode()
    cursor = new_linked_list
    while k > 0:
        cursor.val = head.val
        head = head.next
        k -= 1
        if k != 0:
            cursor.next = ListNode()
            cursor = cursor.next
    return new_linked_list

def print_linked_list(head):
    while head is not None:
        print(head.val)
        head = head.next
        
def reverse_linked_list(head):
    stack = Stack()
    
    # push all elements in stack
    while head is not None:
        stack.push(head.val)
        head = head.next
    
    new_linked_list = ListNode()
    new_head = new_linked_list
    
    while not stack.isEmpty():
        new_head.val = stack.pop()
        if not stack.isEmpty():
            new_head.next = ListNode()
            new_head = new_head.next
    return new_linked_list
        
def skip_k_nodes(head, k):
    while k > 0:
        head = head.next
        k -= 1
    return head

class Stack(object):
    # Reference: https://runestone.academy/runestone/books/published/pythonds/BasicDS/ImplementingaStackinPython.html
    def __init__(self):
         self.items = []
            
    def isEmpty(self):
        return self.items == []
    
    def push(self, item):
        self.items.append(item)
        
    def pop(self):
        return self.items.pop()
    
    def peek(self):
        return self.items[len(self.items)-1]
    
    def size(self):
        return len(self.items)
```
