```python
class Solution(object):
    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """
        # Solution 2. Use hash table (dictionary)
        idx_list = []
        word_count = {}
        for word in words:
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1
        
        if len(words) > 0:
            word_len = len(words[0])
            num_words = len(words)
            len_to_check = word_len * num_words
            for i in range(0, len(s) - len_to_check + 1):
                substr_to_check = s[i:(i + len_to_check)]
                word_count_tmp = {}
                while len(substr_to_check) > 0:
                    target_word = substr_to_check[:word_len]
                    if target_word in word_count:
                        if target_word in word_count_tmp:
                            word_count_tmp[target_word] += 1
                        else:
                            word_count_tmp[target_word] = 1
                        
                        if word_count_tmp[target_word] > word_count[target_word]:
                            break
                        else:
                            substr_to_check = substr_to_check[word_len:]
                    else:
                        break
                if (len(substr_to_check) == 0):
                    idx_list.append(i)
                    
        return idx_list
    
    # Naive solution --> time out
    # idx_list = []
    # if len(words) > 0:
    #     word_len = len(words[0])
    #     num_words = len(words)
    #     len_to_check = word_len * num_words
    #     for i in range(0, len(s) - len_to_check + 1):
    #         substr_to_check = s[i:(i + len_to_check)]
    #         words_remainder = [word for word in words]
    #         while len(words_remainder) > 0:
    #             target_word = substr_to_check[:word_len]
    #             if target_word in words_remainder:
    #                 words_remainder = remove_first_matched_element(words_remainder, target_word)
    #                 substr_to_check = substr_to_check[word_len:]
    #             else:
    #                 break
    #         print(i, len(substr_to_check), (len(words_remainder)))
    #         if (len(substr_to_check) == 0) and (len(words_remainder) == 0):
    #             idx_list.append(i)
    # return idx_list
        

# def remove_first_matched_element(lst, element):
#     new_lst = []
#     matched_flag = False
#     for word in lst:
#         if (word == element) and (not matched_flag):
#             matched_flag = True
#         else:
#             new_lst.append(word)
#     return new_lst
```
