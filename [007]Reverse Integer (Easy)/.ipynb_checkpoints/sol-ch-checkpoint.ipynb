```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        str_x = str(x)
        if x == 0:
            return x
        elif x < 0:
            negative = True
            str_x = str_x[1:]
        else:
            negative = False
            
        i = 0
        end_of_front_zeros = False
        reversed_str_x = str_x[::-1]
        answer = ""
        
        while i < len(reversed_str_x):
            if (reversed_str_x[i] == 0) and (not end_of_front_zeros):
                i += 1
            else:
                end_of_front_zeros = True
                answer = answer + reversed_str_x[i]
                i += 1
                
        if negative:
            answer = "-" + answer
            
        lower_limit = str(-2**31)
        upper_limit = str((2**31) - 1)
        if (compare_int_str(answer, upper_limit)==1) or (compare_int_str(answer, lower_limit)==-1):
            return 0
        else:
            return int(answer)
    
def compare_int_str(s1, s2):
    if s1.startswith('-') and (not s2.startswith('-')):
        return -1
    elif (not s1.startswith('-')) and (s2.startswith('-')):
        return 1
    elif (not s1.startswith('-')) and (not s2.startswith('-')):
        return compare_abs_str(s1, s2)
    else: # two str are negative
        return (-1) * compare_abs_str(s1[1:], s2[1:])

def compare_abs_str(abs_s1, abs_s2):
    # 1: abs_s1 > abs_s2, 0: abs_s1 == abs_s2, -1: abs_s1 < abs_s2
    if len(abs_s1) > len(abs_s2):
        return 1
    elif len(abs_s1) < len(abs_s2):
        return -1
    else:  # two str have the same number of chars
        for i in range(len(abs_s1)):
            s1_digit = int(abs_s1[i])
            s2_digit = int(abs_s2[i])
            if s1_digit > s2_digit:
                return 1
            elif s1_digit < s2_digit:
                return -1
            else:
                continue
    return 0
```
