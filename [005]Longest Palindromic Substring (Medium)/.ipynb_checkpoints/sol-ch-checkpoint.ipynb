```python
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        if len(s) > 1:
            two_char_panlindrome_idx = getTwoConsecutiveCharPanlindrome(s)
            three_char_panlindrome_idx = getThreeCharPanlindrome(s)
            expanded_two_char_panlindrom = expandBasicPanlindrome(s, two_char_panlindrome_idx, two_char=True)
            expanded_three_char_panlindrom = expandBasicPanlindrome(s, three_char_panlindrome_idx, two_char=False)
            merged = expanded_two_char_panlindrom + expanded_three_char_panlindrom
            if len(merged) == 0:
                max_str = s[0]
            else:
                max_str, max_len = argmax(merged)
        else:
            max_str = s
        return max_str
        
def getTwoConsecutiveCharPanlindrome(s):
    idx_lst = []  # (idx, idx+1) has the same char
    for i in range(len(s)-1):
        if s[i] == s[i+1]:
            idx_lst.append(i)
    return idx_lst

def getThreeCharPanlindrome(s):
    idx_lst = []  # (idx - 1, idx, idx + 1) constructs a panlindrome
    for i in range(1, len(s)-1):
        if s[i-1]==s[i+1]:
            idx_lst.append(i)
    return idx_lst

def expandBasicPanlindrome(s, idx_lst, two_char):
    expanded_panlindrome = []
    for idx in idx_lst:
        if two_char:
            left = idx
            right = idx + 1
        else:
            left = idx - 1
            right = idx + 1

        while (left > 0) & (right < len(s) - 1):
            if s[left-1] == s[right+1]:
                left = left - 1
                right = right + 1
            else:
                break
        expanded_panlindrome.append(s[left:(right+1)])
    return expanded_panlindrome

def argmax(lst):
    max_len = 0
    max_str = ""
    for substr in lst:
        if len(substr) > max_len:
            max_len = len(substr)
            max_str = substr
    return max_str, max_len
```
