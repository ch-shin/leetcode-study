```python
INT_MIN = -2**31
INT_MAX = 2 ** 31 - 1

class Solution:
    def myAtoi(self, str: str) -> int:
        str_x = str
        i = 0
        filtered_str = ""
        
        # skip until i meets '-' or digits
        while i < len(str_x):
            if str_x[i] == " ":
                i += 1
            elif str_x[i] == "-":
                filtered_str = filtered_str + str_x[i]
                i += 1
                break
            elif str_x[i] == "+":
                i += 1
                break
            elif str_x[i] in "0123456789":
                break
            else:
                return 0
        
        # skip zeros
        while i < len(str_x):
            if str_x[i] == "0":
                i += 1
            else:
                break
        
        while i < len(str_x):
            if str_x[i] in "0123456789":
                filtered_str = filtered_str + str_x[i]
                i += 1
            else:
                break

        if (filtered_str == "-") or (filtered_str == ""):
            return 0
        elif compare_int_str(filtered_str, INT_MAX.__str__())==1:
            return INT_MAX
        elif compare_int_str(filtered_str, INT_MIN.__str__())==-1:
            return INT_MIN
        else:
            return int(filtered_str)


def compare_int_str(s1, s2):
    if s1.startswith('-') and (not s2.startswith('-')):
        return -1
    elif (not s1.startswith('-')) and (s2.startswith('-')):
        return 1
    elif (not s1.startswith('-')) and (not s2.startswith('-')):
        return compare_abs_str(s1, s2)
    else: # two str are negative
        return (-1) * compare_abs_str(s1[1:], s2[1:])

def compare_abs_str(abs_s1, abs_s2):
    # 1: abs_s1 > abs_s2, 0: abs_s1 == abs_s2, -1: abs_s1 < abs_s2
    if len(abs_s1) > len(abs_s2):
        return 1
    elif len(abs_s1) < len(abs_s2):
        return -1
    else:  # two str have the same number of chars
        for i in range(len(abs_s1)):
            s1_digit = int(abs_s1[i])
            s2_digit = int(abs_s2[i])
            if s1_digit > s2_digit:
                return 1
            elif s1_digit < s2_digit:
                return -1
            else:
                continue
    return 0
```
